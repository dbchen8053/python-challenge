```python
import os
import csv

csvpath = os.path.join('.', 'election_data.csv')

with open(csvpath, newline='') as csvfile:

    csvreader = csv.reader(csvfile, delimiter=',')
    csv_header = next(csvreader)
    
    candidate = []
    
    for row in csvreader:
        candidate.append(row[2])
    vote_total = (len(candidate))
    
    k_counter = 0
    c_counter = 0
    l_counter = 0
    o_counter = 0
    
    for people in candidate:
        if (people == "Khan"):
            k_counter = k_counter + 1
        elif (people == "Correy"):
            c_counter = c_counter + 1
        elif (people == "Li"):
            l_counter = l_counter + 1
        else:
            o_counter = o_counter + 1
    
    k_percent = k_counter/vote_total *100
    c_percent = c_counter/vote_total *100    
    l_percent = l_counter/vote_total *100    
    o_percent = o_counter/vote_total *100
    
    k_vote = k_counter
    c_vote = c_counter
    l_vote = l_counter
    o_vote = o_counter
    
    candidate_percent = [k_percent, c_percent, l_percent, o_percent]
    max_candidate_percent = max(candidate_percent)
    max_candidate_percent_index = candidate_percent.index(max_candidate_percent)
       
    candidate_list =["Khan", "Correy", "Li", "O'Tooley"]
    win_candidate = candidate_list[max_candidate_percent_index]
    
    
output_path = os.path.join('.', 'election_result.txt')
with open(output_path, 'w') as txtfile:
    txtfile.write (f"Election Results" + '\n')
    txtfile.write ("----------------------------" + '\n')
    txtfile.write (f"Total Votes: {vote_total}" + '\n')
    txtfile.write ("----------------------------" + '\n')
    txtfile.write (f"{candidate_list[0]}: {candidate_percent[0]}% ({k_vote})" + '\n')
    txtfile.write (f"{candidate_list[1]}: {candidate_percent[1]}% ({c_vote})" + '\n')
    txtfile.write (f"{candidate_list[2]}: {candidate_percent[2]}% ({l_vote})" + '\n')
    txtfile.write (f"{candidate_list[3]}: {candidate_percent[3]}% ({o_vote})" + '\n')
    txtfile.write ("----------------------------" + '\n')
    txtfile.write (f"Winner: {win_candidate}" + '\n')
    txtfile.write ("----------------------------" + '\n')
    
    
    
```
